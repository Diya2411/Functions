```python
1. Write a Python program to print "Hello, World!"
Answer- # Python 'Hello world' program
          print("Hello world")

2. Write a Python program that displays your name and age.
Answer- # Define name and age
name = "Diya Mishra"
age = 22
# Display output
print("My name is", name)
print("I am", age, "years old")

3. Write code to print all the pre-defined keywords in Python using the keyword
library.
Answer- import keyword
# Get the list of all Python keywords
keywords_list = keyword.kwlist

# Print the keywords
print("Python Keywords:")
for kw in keywords_list:
    print(kw)

4. Write a program that checks if a given word is a Python keyword.
Answer- import keyword
# Take user input
word = input("Enter a word to check: ")

# Check if it's a Python keyword
if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is NOT a Python keyword.")

5. Create a list and tuple in Python, and demonstrate how attempting to change
an element works differently for each.
Answer- # Creating a list
my_list = [1, 2, 3, 4, 5]
print("Original List:", my_list)

# Modifying an element in the list
my_list[2] = 10
print("Modified List:", my_list)

# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)
print("\nOriginal Tuple:", my_tuple)

# Attempting to modify an element in the tuple
try:
    my_tuple[2] = 10  # This will raise an error
except TypeError as e:
    print("Error:", e)

6.  Write a function to demonstrate the behavior of mutable and immutable
arguments.
Answer- def modify_arguments(immutable_arg, mutable_arg):
    # Trying to modify the immutable argument (integer)
    immutable_arg += 10  # This creates a new integer object

    # Trying to modify the mutable argument (list)
    mutable_arg.append(100)  # This modifies the existing list

    print("Inside function:")
    print("Immutable argument:", immutable_arg)
    print("Mutable argument:", mutable_arg)

  # Immutable argument (integer)
    num = 50

  # Mutable argument (list)
    numbers_list = [1, 2, 3]

    print("Before function call:")
    print("Immutable argument:", num)
    print("Mutable argument:", numbers_list)

  # Function call
modify_arguments(num, numbers_list)

print("\nAfter function call:")
print("Immutable argument:", num)

7. Write a program that performs basic arithmetic operations on two user-input
numbers.
Answer- # Taking input from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing basic arithmetic operations
sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2

# Handling division to avoid division by zero error
if num2 != 0:
    quotient = num1 / num2
    remainder = num1 % num2
else:
    quotient = "Undefined (cannot divide by zero)"
    remainder = "Undefined (cannot divide by zero)"

# Displaying the results
print("\nResults:")
print(f"Sum: {sum_result}")
print(f"Difference: {difference}")
print(f"Product: {product}")
print(f"Quotient: {quotient}")
print(f"Remainder: {remainder}")

8. Write a program to demonstrate the use of logical operators.
Answer- # Taking input from the user
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Using logical AND
print("\nLogical AND:")
print(f"Both numbers are positive: {a > 0 and b > 0}")

# Using logical OR
print("\nLogical OR:")
print(f"At least one number is positive: {a > 0 or b > 0}")

# Using logical NOT
print("\nLogical NOT:")
print(f"a is NOT positive: {not (a > 0)}")
print(f"b is NOT positive: {not (b > 0)}")

9. Write a Python program to convert user input from string to integer, float,
and boolean types.
Answer- # Taking user input as a string
user_input = input("Enter a value: ")

# Converting to integer (if possible)
try:
    int_value = int(user_input)
except ValueError:
    int_value = "Cannot convert to integer"

# Converting to float (if possible)
try:
    float_value = float(user_input)
except ValueError:
    float_value = "Cannot convert to float"

# Converting to boolean
bool_value = bool(user_input)  # Empty string is

10. Write code to demonstrate type casting with list elements.
Answer- List after converting to integers: [10, 20, 30, 40, 50]
List after converting to floats: [10.0, 20.0, 30.0, 40.0, 50.0]
List after converting back to strings: ['10', '20', '30', '40', '50']
List after converting to booleans: [True, True, True, True, True]

11.  Write a program that checks if a number is positive, negative, or zero.
Answer- # Program to check if a number is positive, negative, or zero

# Input from user
num = float(input("Enter a number: "))

# Checking conditions
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

12. Write a for loop to print numbers from 1 to 10.
Answer- # Using a for loop to print numbers from 1 to 10
for num in range(1, 11):
    print(num)

13. Write a Python program to find the sum of all even numbers between 1 and 50.
Answer- # Program to find the sum of even numbers between 1 and 50

sum_even = 0  # Initialize sum

# Loop through even numbers from 2 to 50
for num in range(2, 51, 2):
    sum_even += num  # Add even number to sum

print("The sum of all even numbers between 1 and 50 is:", sum_even)

14. Write a program to reverse a string using a while loop.
Answer- # Program to reverse a string using a while loop

# Input from user
string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(string) - 1  # Start from the last character

# Loop to reverse the string
while index >= 0:
    reversed_string += string[index]
    index -= 1  # Move to the previous character

# Output the reversed string
print("Reversed string:", reversed_string)

15.  Write a Python program to calculate the factorial of a number provided by
the user using a while loop.
Answer- # Program to calculate the factorial of a number using a while loop

# Input from user
num = int(input("Enter a number: "))

# Initialize factorial and counter
factorial = 1
i = num

# Calculate factorial using while loop
while i > 0:
    factorial *= i
    i -= 1  # Decrease the counter

# Output the factorial
print(f"The factorial of {num} is: {factorial}")
```
